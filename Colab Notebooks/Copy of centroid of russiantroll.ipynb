```
def average_centroid_dist(data2d,cluster_ar,km_model):
  centroid = PCA(n_components=2).fit_transform(km_model.cluster_centers_)
  tot_dist = 0

  for i in range(cluster_ar.size):
    #print('{},{}'.format(np.asarray(data2d[i])[0],np.asarray(data2d[i])[1]))
    x = np.asarray(data2d[i])[0]
    y = np.asarray(data2d[i])[1]
    cluster = cluster_ar[i]
    dist = (x - centroid[cluster][0])**2 + (y - centroid[cluster][1])**2
    tot_dist += dist ** 0.5
  return tot_dist, cluster_ar.size, tot_dist/cluster_ar.size
```
